```python
import pandas as pd
import numpy as np
import random
from faker import Faker

fake = Faker() # creates instance of faker library 
```

```python

# Define product categories
categories = {
    "Electronics": ["Smartphone", "Laptop", "Tablet", "Smartwatch", "Headphones", "Camera"],
    "Home Appliances": ["Microwave", "Refrigerator", "Washing Machine", "Air Conditioner", "Vacuum Cleaner"],
    "Decor": ["Wall Clock", "Table Lamp", "Cushion", "Vase", "Curtains"],
    "Beauty": ["Lipstick", "Foundation", "Perfume", "Shampoo", "Moisturizer"],
    "Stationery": ["Notebook", "Pen", "Marker", "Eraser", "Stapler"],
    "Wellness": ["Yoga Mat", "Protein Powder", "Vitamins", "Face Mask", "Massage Oil"]
}

# Pre-generate 35 unique suppliers
suppliers = [fake.company() for _ in range(35)]

# Pre-generate 40 unique warehouse locations 
warehouse_locations = [
    "Mumbai, Maharashtra", "Delhi, Delhi", "Bengaluru, Karnataka",
    "Hyderabad, Telangana", "Chennai, Tamil Nadu", "Kolkata, West Bengal",
    "Pune, Maharashtra", "Ahmedabad, Gujarat", "Jaipur, Rajasthan",
    "Lucknow, Uttar Pradesh", "Kanpur, Uttar Pradesh", "Nagpur, Maharashtra",
    "Indore, Madhya Pradesh", "Bhopal, Madhya Pradesh", "Surat, Gujarat",
    "Vadodara, Gujarat", "Patna, Bihar", "Ranchi, Jharkhand",
    "Raipur, Chhattisgarh", "Guwahati, Assam", "Chandigarh, UT",
    "Noida, Uttar Pradesh", "Gurugram, Haryana", "Amritsar, Punjab",
    "Jalandhar, Punjab", "Coimbatore, Tamil Nadu", "Madurai, Tamil Nadu",
    "Thiruvananthapuram, Kerala", "Kochi, Kerala", "Mysuru, Karnataka",
    "Mangalore, Karnataka", "Visakhapatnam, Andhra Pradesh",
    "Vijayawada, Andhra Pradesh", "Bhubaneswar, Odisha", "Cuttack, Odisha",
    "Dehradun, Uttarakhand", "Shimla, Himachal Pradesh",
    "Jammu, J&K", "Srinagar, J&K"
]



# data generation

def generate_product(pid):
    category = random.choice(list(categories.keys()))
    product_base = random.choice(categories[category]) # pick a product (value) from the randomly chosen category(key)
    ## product_base uses the category randomly chosen in the category loc
    
    # Product & supplier
    product_name = f"{fake.company()} {product_base}" 
    supplier = random.choice(suppliers)

    
    # Pricing
    cost_price = round(random.uniform(50, 50000), 2)
    unit_price = round(cost_price * random.uniform(1.1, 1.8), 2) # the margin being 20% to 80%
    
    # Warehouse location 
    warehouse_location = random.choice(warehouse_locations)
    
    return {
        "ProductID": pid,
        "ProductName": product_name,
        "Category": category,
        "Supplier": supplier,
        "CostPrice": cost_price,
        "UnitPrice": unit_price,
        "WarehouseLocation": warehouse_location
    }

# Generate 100 products
products = [generate_product(i+1) for i in range(100)]
df_products = pd.DataFrame(products)
```
